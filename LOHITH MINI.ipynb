```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your CSV file (assuming it's the Iris dataset)
file_path = r'C:\Users\bharath\OneDrive\Desktop\GAMES CSV.csv'
df = pd.read_csv(file_path)

# Display basic statistics
print("Summary Statistics:")
print(df.describe())

# Univariate analysis - Histogram and Box Plot
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['BGMI'], kde=True, bins=20, color='skyblue')
plt.title('Histogram of BGMI')
plt.xlabel('BGMI')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['BGMI'], color='lightcoral')
plt.title('Box Plot of BGMI')
plt.xlabel('BGMI')

plt.tight_layout()
plt.show()

# Multivariate analysis - Pair Plot
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='VALO', palette='Set2')
plt.suptitle('Pair Plot of GAMES CSV Dataset', y=1.02)
plt.show()
```
