```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your CSV file (assuming it's the Iris dataset)
file_path = r'C:\Users\bharath\OneDrive\Desktop\IRIS.csv'
df = pd.read_csv(file_path)

# Display basic statistics
print("Summary Statistics:")
print(df.describe())

# Univariate analysis - Histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['LONGITUDE'], kde=True, bins=20, color='skyblue')
plt.title('Histogram of LONGITUDE')
plt.xlabel('LONGITUDE')
plt.ylabel('Frequency')

# Univariate analysis - Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(x=df['LONGITUDE'], color='lightcoral')
plt.title('Box Plot of LONGITUDE')
plt.xlabel('LONGITUDE')

plt.tight_layout()
plt.show()

# Multivariate analysis - Pair Plot
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='ALTITUDE', palette='Set2')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

# Multivariate analysis - Pair Plot
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='ALTITUDE', palette='Set2')  # Assuming 'SPECIES' is a categorical column
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your CSV file (replace 'your_file.csv' with the actual file path)
file_path = r'C:\Users\bharath\OneDrive\Desktop\IRIS.csv'
df = pd.read_csv(file_path)

# Choose the variable for univariate analysis (replace 'COLUMN_NAME' with the actual column name)
variable_of_interest = 'LONGITUDE'

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df[variable_of_interest], kde=True, bins=20, color='skyblue')
plt.title(f'Histogram of {variable_of_interest}')
plt.xlabel(variable_of_interest)
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df[variable_of_interest], color='lightcoral')
plt.title(f'Box Plot of {variable_of_interest}')
plt.xlabel(variable_of_interest)
plt.show()

# Kernel Density Plot (KDE)
plt.figure(figsize=(8, 6))
sns.kdeplot(df[variable_of_interest], shade=True, color='salmon')
plt.title(f'Kernel Density Plot of {variable_of_interest}')
plt.xlabel(variable_of_interest)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = r'C:\Users\bharath\OneDrive\Desktop\200.csv'
df1 = pd.read_csv(file_path)

# Fill missing values
df1['bp.2s'].fillna(df1['bp.2s'].mean(), inplace=True)
df1['bp.2d'].fillna(df1['bp.2d'].mean(), inplace=True)
df1['glyhb'].fillna(df1['glyhb'].mean(), inplace=True)
df1['chol'].fillna(df1['chol'].mean(), inplace=True)
df1['hdl'].fillna(df1['hdl'].mean(), inplace=True)
df1['ratio'].fillna(df1['ratio'].mean(), inplace=True)
df1['height'].fillna(df1['height'].mean(), inplace=True)
df1['weight'].fillna(df1['weight'].mean(), inplace=True)
df1['bp.1d'].fillna(df1['bp.1d'].mean(), inplace=True)
df1['time.ppn'].fillna(df1['time.ppn'].mean(), inplace=True)
df1['bp.1s'].fillna(df1['bp.1s'].mean(), inplace=True)


# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df1.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values Heatmap')
plt.show()
```
